```python
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
import pandas as pd
import matplotlib.pyplot as plt

data = [
    ("Fluconazole", "C1=CC(=C(C=C1F)F)C(CN2C=NC=N2)(CN3C=NC=N3)O"),
    ("Ketoconazole", "CC(=O)N1CCN(CC1)C2=CC=C(C=C2)OC[C@@H]3CO[C@@](O3)(CN4C=CN=C4)C5=C(C=C(C=C5)Cl)Cl"),
    ("Itraconazole", "CCC(C)N1C(=O)N(C=N1)C2=CC=C(C=C2)N3CCN(CC3)C4=CC=C(C=C4)OC[C@H]5CO[C@](O5)(CN6C=NC=N6)C7=C(C=C(C=C7)Cl)Cl"),
    ("Resveratrol", "C1=CC(=CC=C1/C=C/C2=CC(=CC(=C2)O)O)O"),
    ("Caffeine", "CN1C=NC2=C1C(=O)N(C(=O)N2C)C"),
    ("Diazepam", "CN1C(=O)CN=C(C2=C1C=CC(=C2)Cl)C3=CC=CC=C3"),
    ("Fluoxetine", "CNCCC(C1=CC=CC=C1)OC2=CC=C(C=C2)C(F)(F)F")
]
```

```python
for name, smiles in data:
    mol = Chem.MolFromSmiles(smiles)
    print(name, "OK" if mol else "FAILED")
```

```python
rows = []

for name, smiles in data:
    mol = Chem.MolFromSmiles(smiles)
    rows.append({
        "Name": name,
        "MW": Descriptors.MolWt(mol),
        "LogP": Descriptors.MolLogP(mol),
        "TPSA": Descriptors.TPSA(mol),
        "HBD": Descriptors.NumHDonors(mol),
        "HBA": Descriptors.NumHAcceptors(mol),
        "RotB": Lipinski.NumRotatableBonds(mol),
        "HeavyAtoms": Descriptors.HeavyAtomCount(mol),
        "AromaticRings": Lipinski.NumAromaticRings(mol),
        "FractionCSP3": Descriptors.FractionCSP3(mol),
        "MolMR": Descriptors.MolMR(mol),
        "FormalCharge": Chem.GetFormalCharge(mol)
    })

df = pd.DataFrame(rows)
df
```

### Comparative Analysis for Fluconazole

**Raw Comparison Notes:**

- TPSA is relatively high at 81.65, suggesting greater polarity compared to compounds typically optimized for CNS penetration.

- MW is 306.3 g/mol, within the range for oral bioavailability, and lower than several other compounds in the set.

- FCZ has 1 HBD and 5 HBA, consistent with drug-like properties, but still indicative of a molecule capable of significant hydrogen bonding.

- LogP is 0.74, indicating that FCZ is relatively hydrophilic and less lipophilic than most CNS-active compounds.

- FCZ has 5 rotatable bonds, suggesting moderate flexibility relative to other compounds in the set.

**Summary:**

- Fluconazole displays a physicochemical profile that aligns more closely with polar, systemically active drugs than with CNS-optimized compounds.

- Its relatively high TPSA and low LogP suggest that passive diffusion across the blood-brain barrier may be limited.

- Although FCZ meets general drug-likeness criteria, its polarity and hydrogen bonding capacity distinguish it from classical CNS-active drugs, which are typically more lipophilic and less polar.

- Taken together, these properties suggest that fluconazole is unlikely to rely solely on passive permeability to exert neuroprotective effects.

- This raises the possibility that its observed activity may involve alternative mechanisms, such as transporter-mediated uptake or indirect modulation of biological pathways.

```python
def group(name):
    if name in ["Fluconazole", "Ketoconazole", "Itraconazole"]:
        return "Azole"
    elif name in ["Caffeine", "Diazepam", "Fluoxetine"]:
        return "CNS"
    elif name == "Resveratrol":
        return "Natural"
    else:
        return "Other"

df["Group"] = df["Name"].apply(group)
df
```

```python
plt.figure(figsize=(7,5))

colors = {"Azole":"blue", "CNS":"red", "Natural":"green", "Other":"gray"}

for g, d in df.groupby("Group"):
    plt.scatter(d["LogP"], d["TPSA"], label=g, color=colors.get(g, "gray"))

for i, txt in enumerate(df["Name"]):
    plt.annotate(txt, (df["LogP"][i], df["TPSA"][i]))

plt.xlabel("LogP")
plt.ylabel("TPSA")
plt.title("LogP vs TPSA (Grouped)")
plt.legend()
plt.show()
```

```python
df["BBB_LogP_ok"] = df["LogP"].between(2, 4)
df["BBB_TPSA_ok"] = df["TPSA"] < 70

df[["Name","LogP","TPSA","BBB_LogP_ok","BBB_TPSA_ok"]]
```

### CNS-Likeness Check

- Most CNS-active compounds fall within a rough LogP range of 2–4 and TPSA below ~70 Å².

- Fluconazole does not meet either criterion (LogP ~0.74; TPSA ~81.65), placing it outside typical CNS-optimized space.

- In the grouped plot, fluconazole clusters with azoles rather than CNS compounds.

- Together, these observations support the conclusion that fluconazole’s neuroprotective effects are unlikely to be driven by passive BBB diffusion alone.

```python
plt.figure(figsize=(7,5))

for g, d in df.groupby("Group"):
    plt.scatter(d["LogP"], d["FractionCSP3"], label=g)

for i, txt in enumerate(df["Name"]):
    plt.annotate(txt, (df["LogP"][i], df["FractionCSP3"][i]))

plt.xlabel("LogP")
plt.ylabel("FractionCSP3")
plt.title("LogP vs FractionCSP3 (3D Character)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7,5))

for g, d in df.groupby("Group"):
    plt.scatter(d["TPSA"], d["RotB"], label=g)

for i, txt in enumerate(df["Name"]):
    plt.annotate(txt, (df["TPSA"][i], df["RotB"][i]))

plt.xlabel("TPSA")
plt.ylabel("Rotatable Bonds")
plt.title("TPSA vs Rotatable Bonds (Flexibility vs Polarity)")
plt.legend()
plt.show()
```

### Additional Structural Context

- In the LogP vs FractionCSP3 plot, fluconazole shows slightly more 3D character than other azoles but remains outside the typical range of CNS-active compounds.

- In the TPSA vs Rotatable Bonds plot, fluconazole combines relatively high polarity with moderate flexibility, a profile that is not typically associated with efficient BBB penetration.

- Together, these plots reinforce the conclusion that fluconazole does not structurally resemble CNS-optimized compounds, despite its observed neuroprotective effects.
